```python
import geopandas as gpd
import pandas as pd
from arcgis.gis import GIS
from shapely import geometry
from shapely.geometry import Point, Polygon
import numpy as np
import pyproj as pj
import folium
import folium.plugins as plugins
from folium.plugins import MarkerCluster
import pyodbc
import math
sql_conn = pyodbc.connect('DRIVER={ODBC Driver 13 for SQL Server};'
    'SERVER=L1179\L1179;DATABASE=OZ_Global_Project_Gen;Trusted_Connection=yes') 
```

```python
query = '''
SELECT * FROM Project_Monitor_Details
'''
dfDetails = pd.read_sql(query, sql_conn)
dfDetails.head()
```

```python
query = '''
SELECT * FROM Project_Monitor_PListing
'''
dfPListing = pd.read_sql(query, sql_conn)
dfPListing.head()
```

```python
df = pd.merge(dfDetails, dfPListing,how='left',left_on=['Project_ID'],right_on=['Project_ID'])
df['Geometry'] = [Point(x,y) for x,y in  zip(df['Project_Lon'],df['Project_Lat'])]
df['OZ_Monitor_Status'] = df['OZ_Monitor_Status'].str.upper()
df = gpd.GeoDataFrame(df,geometry = 'Geometry',crs="+init=epsg:4326")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
dfNoReject = df.loc[df['OZ_Monitor_Status'] != 'REJECT']
```

```python
dfNoReject_NoCollar = dfNoReject[dfNoReject['Project_Lat'].isnull()]
dfNoReject_Collar  = dfNoReject[~dfNoReject['Project_Lat'].isnull()]
```

```python
dfNoReject_NoCollar
```

```python
dfNoReject_Collar.iloc[0]
```

```python
m = folium.Map(tiles = 'Stamen Terrain')
marker_cluster = MarkerCluster().add_to(m)
for index, row in dfNoReject_Collar.iterrows():
    folium.Marker(
            [row['Geometry'].y,row['Geometry'].x],
            #color=colordict[traffic_q],
            fill_color='#3186cc',
            fill=True,
            fill_opacity=0.7
            ).add_to(marker_cluster)
    try:
        message = ('Project ID:' + str(row['Project_ID']) + '<br>' 
                   'Commodities:' + row['Deposit_Commodities'])
    except:
        message = ('Project ID:' + str(row['Project_ID']))
    folium.Popup(message).add_to(marker_cluster)
m
```

```python
style = {'description_width': '150px'}
layout1 = {'width':'600px'}

class details_input():
    def __init__(self):
        #self.Reporting_Date = Text(description = 'Reporting Date',layout=Layout(width='50%', height='80px'))
        self.Reporting_Date = widgets.DatePicker(description = 'Reporting Date',style=style)
        self.OZ_Project_Dir = widgets.Text(description = 'Project Directory',style=style,layout=layout1)
        self.OZ_Monitor_Status = widgets.Dropdown(options=['Active Project', 'Advance', 'Monitor','Reject'],
                                                  value='Active Project',description = 'Monitor Status',style=style)   
        self.OZ_Monitor_Summary = widgets.Textarea(description = 'Summary',style=style,layout=layout1)
        self.SNL_Weblink = widgets.Text(description = 'Web Link',style=style,layout=layout1)
        self.Project_Lat = widgets.Text(description = 'Latitude',style=style)
        self.Project_Lon = widgets.Text(description = 'Longitude',style=style)
        self.Deposit_Model = widgets.Text(description = 'Deposit Model',style=style)
        self.Deposit_Commodities = widgets.Text(description = 'Deposit Commodities',style=style,layout=layout1)
        self.Deposit_Mining_Method = widgets.Text(description = 'Mining Method',style=style)
        self.Global_Resource_Text = widgets.Textarea(description = 'Global Resource',style=style,layout=layout1)
        self.Project_Dev_Stage = widgets.Text(description = 'Development Stage',style=style)
        self.Previous_Reporting_Geologist = widgets.Text(description = 'Previous Geologist',style=style)
        self.Insert_Button = widgets.Button(value=False,description='Insert',icon='check')
        #self.Reporting_Date.on_submit(self.handle_submit)

    def handle_submit(self, text):
        self.v = text.value
        return self.v

class list_input():
    def __init__(self):
        self.Project_Name = widgets.Text(description = 'Project Name',style=style)
        self.Project_Owner = widgets.Text(description = 'Project Owner',style=style)
        self.Project_LocCountry = widgets.Text(description = 'Country',style=style)  
        self.Project_LocState = widgets.Text(description = 'State',style=style)
        self.Reporting_Geo = widgets.Text(description = 'Reporting Geologist',style=style)
    
def on_insert_clicked(b):
    try:
        query = 'SELECT MAX(Project_ID) AS MAX_PROJECT_ID FROM Project_Monitor_Details'
        dfProject_ID = pd.read_sql(query, sql_conn)
        Project_ID = int(np.array(dfProject_ID.iloc[0].values)[0]) + 1
        Reporting_Date = f.Reporting_Date.value     
        OZ_Project_Dir = f.OZ_Project_Dir.value    
        OZ_Monitor_Status = f.OZ_Monitor_Status.value    
        OZ_Monitor_Summary = f.OZ_Monitor_Summary.value 
        SNL_Weblink = f.SNL_Weblink.value 
        Project_Lat = f.Project_Lat.value  
        Project_Lon = f.Project_Lon.value 
        Deposit_Model = f.Deposit_Model.value 
        Deposit_Commodities = f.Deposit_Commodities.value 
        Deposit_Mining_Method = f.Deposit_Mining_Method.value 
        Global_Resource_Text = f.Global_Resource_Text.value 
        Project_Dev_Stage = f.Project_Dev_Stage.value 
        Previous_Reporting_Geologist = f.Previous_Reporting_Geologist.value 
        
        query = '''
                    INSERT INTO Project_Monitor_Details (Project_ID,Reporting_Date,OZ_Project_Dir,OZ_Monitor_Status,
                                                         OZ_Monitor_Summary, SNL_Weblink, Project_Lat, Project_Lon,
                                                         Deposit_Model, Deposit_Commodities, Deposit_Mining_Method, Global_Resource_Text,
                                                         Project_Dev_Stage, Previous_Reporting_Geologist)
                    VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?)
                '''
        values = (Project_ID, Reporting_Date, OZ_Project_Dir, OZ_Monitor_Status,
                  OZ_Monitor_Summary, SNL_Weblink, Project_Lat, Project_Lon,
                  Deposit_Model, Deposit_Commodities, Deposit_Mining_Method, Global_Resource_Text,
                  Project_Dev_Stage, Previous_Reporting_Geologist)
        cursor = sql_conn.cursor()
        cursor.execute(query,values)
        sql_conn.commit()
        print('Data inserted')
    except Exception as e:
        print ('Error! Code: {c}, Message, {m}'.format(c = type(e).__name__, m = str(e)))


print("OZ Global Data Entry")
f = details_input()
g = list_input()
tab1 = VBox(children=[f.Reporting_Date, f.OZ_Project_Dir, f.OZ_Monitor_Status, 
                      f.OZ_Monitor_Summary, f.SNL_Weblink, f.Project_Lat, 
                      f.Project_Lon, f.Deposit_Model, f.Deposit_Commodities,
                      f.Deposit_Mining_Method, f.Global_Resource_Text, f.Project_Dev_Stage,
                      f.Previous_Reporting_Geologist, f.Insert_Button])
tab2 = VBox(children=[g.Project_Name, g.Project_Owner, g.Project_LocCountry, g.Project_LocState, g.Reporting_Geo])
tab = widgets.Tab(children=[tab1,tab2])
tab.set_title(0, 'Project Details')
tab.set_title(1, 'Project List')
display(tab)

f.Insert_Button.on_click(on_insert_clicked)
```
